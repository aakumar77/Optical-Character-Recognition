```python
import cv2
img_file='Images/data.jpg'
img_scalar =cv2.imread(img_file)
#cv2.imshow("Image",img)
#cv2.waitKey(0)  #without waitKey VScode cant display the image, crashing the window
```

```python
import matplotlib.pyplot as plt
def display(image):
    img_read=plt.imread(image)
    height,width,depth= img_read.shape
    
    dpi=80
    figsize=height/float(dpi), width/float(dpi)
    fig=plt.figure(figsize=figsize)
    
    ax=fig.add_axes([0,0,1,1])
    ax.axis('off')
    
    ax.imshow(img_read, cmap='grey')
    plt.show()
    
display(img_file)
```

```python
#Inverting the colour
inverted_image=cv2.bitwise_not(img_scalar)
cv2.imwrite("Images/data_inverted.jpg", inverted_image)
display("Images/data_inverted.jpg")
```

```python
#Greyscaling the image
def greyscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
grey_image=greyscale(inverted_image)
cv2.imwrite("Images/data_greyscale.jpg",grey_image)
```
